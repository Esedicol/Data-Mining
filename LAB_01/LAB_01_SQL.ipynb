# **Step 8**

**Task: Cleanse	 the	 content	 of	 the	 field	 "year".	 What	 years	 are represented?	 Describe	 any	 errors	 that	 you	 see	 in	 data.	 How many	rows	are	affected?**

Based on the unique values below we can identify that:
- Only years from 2015 to 2018 is present in the column
- The year is displayed in two different ways:<br>
    1) one is displayed fully (i.e. 2018) <br>
    2) the other format only displays the last two digit (i.e. 17 instead of 2017)<br>

```python
# Show all unique values to see if there are any errors
year = sqlDataFrame.groupby('year')['year'].nunique()
print(year, '\n')

# Check for NaN
print("Number of NaN's: ", sqlDataFrame['year'].astype(str).str.contains('nan').sum())
```

```python
# We cast the object of 'year' into a string so we can find its length and then we iterate through each value
for x in sqlDataFrame['year'].astype(str):
    if len(x) == 2:
        
        # This works so we will use this later
        updatedYear = '20' + x
        print(x, " -- ", updatedYear)
```

```python
# To Apply changes to our DataFrame we will use pandas.apply and implement a lambda function

# First we cast the column as a String so we can iterate through each values to get lenght, in our case either 2 or 4
sqlDataFrame['year'] = sqlDataFrame['year'].astype(str) 

sqlDataFrame['year'] = sqlDataFrame.apply(
    
    # Add '20' if the lenght == 2 (17 or 18)
    lambda row : ('20' + row['year']) if len(row['year']) == 2 
    
    # Else do nothing. The 'axis=1' allows us to make the changes on one column only
    else row['year'], axis=1
)
```

# **Step 9**

**Task: Cleanse	 the	 content	 of	 the	 field	 "type".	 What	 avocado type	 are	 represented?	 Describe	 any	 errors	 that	 you	 see.	 How many	rows	are	affected?**

The column 'type' only contains one type of Avocado 'Convnetional'. But there exist errors, we can conclude from the list below that some are not capitalize.


```python
# List out all the unique values to scan out errors
type = sqlDataFrame.groupby('type')['type'].nunique()
print(type)
```

```python
# We will use the 's.str.title' to convert values into capitalize values
sqlDataFrame['type'] = sqlDataFrame['type'].astype(str).str.strip().str.title()

# To prove the method above worked we will print out all unique values and we should only get one 
print(type)
```

```python
# Use the same method from section 2 to reformat 'Date' into %d-%m-%Y
newDate = pd.to_datetime(sqlDataFrame['Date'], format="%d-%m-%Y", errors='coerce')

# The function below will switch the first(%Y) and last(%d)
def switch(input):

    input[0], input[-1] = input[-1], input[0]

    newDate = '-'.join(input)
    return newDate

index = sqlDataFrame[newDate.isnull() == True].index
sqlDataFrame['fValue'] = ''

sqlDataFrame['Date'] = sqlDataFrame['Date'].astype(str)
for i in index:
    xDate = sqlDataFrame['Date'][i]
    s = xDate.split('-')

    if len(s) == 3:
        sqlDataFrame.at[i, 'fValue'] = switch(s)  

sqlDataFrame['Date'] = newDate
sqlDataFrame['Date'] = sqlDataFrame['Date'].fillna(sqlDataFrame['fValue'])

# Drop the 'fValue' as we dont need it anymore
sqlDataFrame.drop('fValue', axis=1, inplace=True)
sqlDataFrame['Date'].head()
```
