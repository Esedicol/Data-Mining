```python
import numpy as np 
import pandas as pd

data_csv = pd.read_csv('BSCY4_csv.csv', delimiter=',')
data_sql = pd.read_csv('BSCY4_sql.csv', delimiter=',')

# Before we merge both DataFrame we need to make sure:
# 1) All Columns have the same type
# 2) Drop Unecessary Columns
# 3) All Columns have the same name
```

```python
# Change type
data_csv['AveragePrice'] = data_csv['AveragePrice'].astype(np.float) 
data_csv['year'] = data_csv['year'].astype(np.int64) 

# Cahnge type to DateTime
data_csv['Date'] = pd.to_datetime(data_csv['Date'])

# Drop Columns
del data_csv['Unnamed: 0']
del data_csv['Unnamed: 0.1']

# Rename Columns
data_csv.columns = ['Date', 'AveragePrice', 'TotalValue', 'c4046', 'c4225', 'c4770', 'TotalBags', 'SmallBags', 'LargeBags', 'XLargeBags', 'type', 'year', 'region']


data_csv.dtypes
```

```python
# Change Type
data_sql['AveragePrice'] = data_sql['AveragePrice'].astype(np.float) 

# Cahnge type to DateTime
data_sql['Date'] = pd.to_datetime(data_sql['Date'])

# Drop Columns
del data_sql['Unnamed: 0']

# Rename Columns
data_sql.columns = ['Date', 'AveragePrice', 'TotalValue', 'c4046', 'c4225', 'c4770', 'TotalBags', 'SmallBags', 'LargeBags', 'XLargeBags', 'type', 'year', 'region']

data_sql.dtypes
```
